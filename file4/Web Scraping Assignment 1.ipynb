### Question 1 - Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
#Send get request to the webpage server to get source code of the page
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
#Now I will extract all the header tags from the page using the for command:
headers = [] #empty list to store data
for i in soup.find_all('span', class_="mw-headline"):
    headers.append(i.text)
    
headers
```

```python
# Making the dataframe
import pandas as pd
df = pd.DataFrame({'Headers':headers})
df
```

### Question 2 - Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
#Send get request to the webpage server to get source code of the page
page_2 = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page_2
```

```python
soup_2 = BeautifulSoup(page_2.content)
soup_2
```

```python
#Extract all the name tags from the page using the for command:
president_name = [] #empty list to store data
for i in soup_2.find_all('div', class_="presidentListing"):
    president_name.append(i.text.split('(')[0])
    
president_name
```

```python
# I don't want the newline character \n to show, so I will use a strip()

president_name = [] #empty list to store data
for i in soup_2.find_all('div', class_="presidentListing"):
    president_name.append(i.text.split('(')[0].strip())

president_name
```

```python
# Next, extract the tags for term of office
term = [] #empty list to store data
for i in soup_2.find_all('span', class_="terms"):
    term.append(i.text)

term
```

```python
# I am only getting Term of office as an output. To get the actual years, I will add next_sibling.strip()

term = [] #empty list to store data 
for i in soup_2.find_all('span', class_="terms"):
    term.append(i.next_sibling.strip())
    
term
```

```python
# Making dataframe
df_2 = pd.DataFrame({'President':president_name,'Term of Office':term})
df_2
```

##### Starting with a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python

#Send get request to the webpage server to get source code of the page
page_3 = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page_3
```

```python
soup_3 = BeautifulSoup(page_3.content)
soup_3
```

```python
#Extract all the teams from the page using the for command:

ODI_Team = [] #empty list to store data
for i in soup_3.find_all('span', class_="u-show-phablet"):
    ODI_Team.append(i.text)
    
ODI_Team = ODI_Team[:10] #Limit the result to TOP 10 rankings
    
ODI_Team
```

##### The number 1 team is in a different class that teams 2 - 10. I will scrape separately and merge both lists together to form a single list

```python
# Next we extract matches

ODI_Match1a = []
for i in soup_3.find_all('td', class_="rankings-block__banner--matches"):
    ODI_Match1a.append(i.text)

ODI_Match1 = []
for i in soup_3.find_all('td', class_="table-body__cell u-center-text"):
    ODI_Match1.append(i.text)

ODI_Match1 = ODI_Match1[0::2][:9]

ODI_Matches = ODI_Match1a + ODI_Match1  # Merge both lists together

ODI_Matches
```

```python
# Next we extract points

ODI_Point1a = []
for i in soup_3.find_all('td', class_="rankings-block__banner--points"):
    ODI_Point1a.append(i.text)

ODI_Point1 = []
for i in soup_3.find_all('td', class_="table-body__cell u-center-text"):
    ODI_Point1.append(i.text)

ODI_Point1 = ODI_Point1[1::2][:9]

ODI_Points = ODI_Point1a + ODI_Point1  # Merge both lists together

ODI_Points
```

```python
# Next we extract rating

ODI_Rating1a = []
for i in soup_3.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    ODI_Rating1a.append(i.text.strip())

ODI_Rating1 = []
for i in soup_3.find_all('td', class_="table-body__cell u-text-right rating"):
    ODI_Rating1.append(i.text)

ODI_Rating1 = ODI_Rating1[:9]

ODI_Ratings = ODI_Rating1a + ODI_Rating1  # Merge both lists together

ODI_Ratings
```

```python
print(len(ODI_Team),len(ODI_Matches),len(ODI_Points),len(ODI_Ratings))
```

```python
# Making dataframe
import pandas as pd
df_3 = pd.DataFrame({'Team':ODI_Team,'Matches':ODI_Matches,'Points':ODI_Points,'Rating':ODI_Ratings})
df_3
```

### Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
#### i) Headline
#### ii) Time
#### iii) News Link

```python
#Send get request to the webpage server to get source code of the page
page_5 = requests.get('https://www.cnbc.com/world/?region=world')
page_5
```

```python
soup_5 = BeautifulSoup(page_5.content)
soup_5
```

```python
#Next I will extract headlines
featured = [] # empty list for storing data

for i in soup_5.find_all('h2', class_="FeaturedCard-packagedCardTitle"):
    featured.append(i.text)

headlineA = [] # empty list for storing data

for i in soup_5.find_all('div', class_="SecondaryCard-headline"):
    headlineA.append(i.text)
    
headlineB = [] # empty list for storing data

for i in soup_5.find_all('div', class_="RiverHeadline-headline RiverHeadline-hasThumbnail"):
    headlineB.append(i.text)
headlines = featured + headlineA + headlineB

headlines
```

### Question 6 - Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
#### i) Paper Title
#### ii) Authors
#### iii) Published Date
#### iv) Paper URL

```python
#Send get request to the webpage server to get source code of the page
page_6 = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page_6
```

```python
soup_6 = BeautifulSoup(page_6.content)
soup_6
```

```python
#Next I will extract paper titles
titles = [] # empty list for storing data

for i in soup_6.find_all('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    titles.append(i.text)
    
titles
```

```python
#Next I will extract authors
authors = [] # empty list for storing data

for i in soup_6.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
    
authors
```

```python
#Next I will extract published date
date = [] # empty list for storing data

for i in soup_6.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)
    
date
```

```python
#Finally extract paper URLs

link = [] # empty list for storing data

for i in soup_6.find_all('a' ,class_="sc-5smygv-0 fIXTHm"):
    link.append(i.get('href'))
    
link
```

```python
# Making dataframe
import pandas as pd
df_6 = pd.DataFrame({'Paper Title':titles,'Authors':authors,'Published Date':date,'Paper URL':link})
df_6
```

### Question 7 - Write a python program to scrape mentioned details from dineout.co.in and make data frame-
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
#### v) Image URL

```python
#Send get request to the webpage server to get source code of the page
page_7 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page_7
```

```python
soup_7 = BeautifulSoup(page_7.content)
soup_7
```

```python
#Next I will extract restaurant names
restaurant = [] # empty list for storing data

for i in soup_7.find_all('a', class_="restnt-name ellipsis"):
    restaurant.append(i.text)
    
restaurant
```

```python
#Next I will extract cuisine

cuisine = [] # empty list for storing data

for i in soup_7.find_all('span' ,class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
    
cuisine
```

```python
# Next I will extract location
location = [] # empty list for storing data

for i in soup_7.find_all('div' ,class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

```python
# Next I will extract ratings

rating = [] # empty list for storing data

for i in soup_7.find_all('div' ,class_="restnt-rating rating-4"):
    rating.append(i.text)
    
rating
```

```python
#Finally extract image URLs

images = [] # empty list for storing data

for i in soup_7.find_all('img' ,class_="no-img"):
    images.append(i.get('data-src'))
    
images
```

```python
# Making dataframe
import pandas as pd
df_7 = pd.DataFrame({'Restaurant':restaurant,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Images_url':images})
df_7
```
